```python
course = course_list[2]

module = course.get_modules()

for m in module:
    print(f"  Module_id: {m.id}")
    print(f"  Module: {m.name}")
    module_items = m.get_module_items()
    for item in module_items:
        print(f"    - Item: {item.title} ({item.type})")
```

```python
# Extracting files from one module [1 = P1W1] (to later create functions for repeatibility)

data = []

sample_m = module[1] # to get the second module, P1W1

items = sample_m.get_module_items()

for i in items:
    print(i.type)
```

```python
sample_m = module[1]

items = sample_m.get_module_items()

for i in items:
    if i.type == "Page":
        page_url = i.page_url
        page_obj = course.get_page(page_url)

        # Parse the page's HTML content
        soup = BeautifulSoup(page_obj.body, 'html.parser')

        # Look for Google Slides URLs in <iframe> and <a> tags
        slide_links = []
        for iframe in soup.find_all('iframe'):
            src = iframe.get('src', '')
            if 'docs.google.com/presentation' in src:
                slide_links.append(src)
        
        for a_tag in soup.find_all('a'):
            href = a_tag.get('href', '')
            if 'docs.google.com/presentation' in href:
                slide_links.append(href)
        
        if slide_links:
            print(f"Found Google Slides links on page '{page_obj.title}':")
            for link in slide_links:
                print(f"  --> {link}")
```

```python
# Extracting files from all modules

items_list = []

for m in module:
    items_list.append(m.get_module_items())

for item in items_list:
    for i in item:
        print(i.type)
```

```python
# function for extracting html object data into a list

filter_out = ["", "\xa0", " \xa0", "\n", "© 2025 The Knowledge House | All Rights Reserved", "© 2025 The Knowledge House | All Rights Reserved "]

def extract_html(new_list, html_data):
    '''
        Input: list and html-formatted data
        Output: list of lists populated with extracted text, links, and link-titles from each module
    '''
    soup = BeautifulSoup(html_data, "html.parser")
    new_list = soup.get_text().splitlines()

    for item in new_list:
        if item in filter_out:
            new_list.pop(new_list.index(item))

    # In this part, link titles and urls will be added to data list *   *   *
    '''
    links = soup.find_all("a")
    for link in links:
        href = link.get("href")
        text = link.get_text()
        print(f"Link: {href}; \nText: {text}\n")
    '''
    return(new_list)
```

```python
# Building up the full data file...

f_data = []

for item in items_list:
    for i in items:
        if i.type == "Page":
            page_id = i.page_url
            page_obj = course.get_page(page_id)
            html_body = page_obj.body
            f_data.append(extract_html(f_data, html_body))
```

```python
# Beginning to build up the data, with smaller lists (starting with page text)

test_list = []

for i in items:
    if i.type == "Page":
        page_id = i.page_url
        print(page_id)
        print("-" * 15)
        page_obj = course.get_page(page_id)
        html_body = page_obj.body
        test_list.append(extract_html(test_list, html_body))
```

```python

def consolidate_list_of_lists(test_list):
    sub_data = []
    i = 0
    ind = len(test_list)
    while i < ind:
        if test_list[i] not in filter_out:
            sub_data.append(" ".join(test_list[i]))
        i += 1

    return(sub_data)
```

```python
data = consolidate_list_of_lists(test_list)

for d in data:
    print(d)
```

```python
f_data = consolidate_list_of_lists(f_data)

f_data
```

```python
short_sentences = []
for f in f_data:
    short_sentences += re.split(r'(?<=[.!?])\s+', f)

short_sentences
```

```python
unique_sents = []
seen_sents = set()

for sent in short_sentences:
    if sent not in seen_sents:
        unique_sents.append(sent)
        seen_sents.add(sent)
    else:
        continue

unique_sents
```

```python
# Do short_sentences have a lot of duplicates???

sentences = {"sentences": short_sentences} # instead of unique_sents

df = pd.DataFrame(sentences)

df.to_csv("data_ds_A.csv", index=False)
```
